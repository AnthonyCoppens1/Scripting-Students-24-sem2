```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
response = requests.get(url)
html = response.text
soup = BeautifulSoup(html)
#print(soup) --> still the same
#print(soup.prettify())
print(soup.title)

print(soup.title.get_text())
print(soup.get_text()[0:500].split())
print(soup.get_text().count("Elephant"))
print(soup.get_text().count("elephant"))

h3_tags = soup.find("h3")
print(h3_tags)

h3_tags = soup.find_all("h3")
print(h3_tags)

for element in h3_tags:
    print(element.get_text())
```

```python
#get all the languages for elephant
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Elephant"
response = requests.get(url)
soup = BeautifulSoup(response.text)

li_tags = soup.find_all("li",{'class':"interlanguage-link"})

for tag in li_tags:
    print(f"{tag.a["title"]} --> {tag.get_text()}")
    #print(tag.a["title"], "-->", tag.get_text()) --> other way of writing it.
```

## More images and regex

```python
import requests, re
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/Soup"
r = requests.get(url)
soup = BeautifulSoup(r.text)

imgTags = soup.find_all('img') #all img tags
sources = [img['src'] for img in imgTags]

result = re.findall(r'(/\w+[.](jpg|gif|png))', str(sources))

for url in result:
    print(url[0])
```

```python
#downloading every single image from weather website
#https://www.meteoblue.com/nl/weer/webcams/barcelona_spanje_3128760

import requests
from bs4 import BeautifulSoup
import os

url = "https://www.meteoblue.com/nl/weer/webcams/barcelona_spanje_3128760"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
images = soup.find_all('img')

#os.mkdir(f"{os.getcwd()}/Images") --> add this only 1 time to immediately create a folder and add all files/images there

count = 0
for img in images:
    url = img["src"]
    filename = f"Images/Barcelona_{count}.jpg"
    response = requests.get(url, allow_redirects=True)
    with open(filename, 'wb') as file:
        file.write(response.content)
    
    count += 1
```
